```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import fastf1.plotting as f1plt
```

```python
# so we have two methods , one is get session which lets us get any session by its year, race(country or number), format
# The other one is get event so we will try and see both of them
import fastf1 as f1


bahrain_session= f1.get_session(2022,1,'R')
bahrain_event= f1.get_event(2022,1)
```

```python
bahrain_session.load(telemetry=True,laps=True,weather=True)
```

```python
laps = bahrain_session.laps # -> 20 drivers and bahrain as 57 laps

laps
```

```python
total_laps_by_each = laps[['Driver','LapNumber']]
total_laps_by_each['Driver'].value_counts()
```

```python
results = bahrain_session.results

results
```

```python
standings = results[['Abbreviation', 'Position']]
standings.rename(columns={'Abbreviation' : 'Driver'})
```

```python
import pandas as pd
grouped =total_laps_by_each['Driver'].value_counts()

finals = pd.concat([standings,grouped])
finals
```

```python

total_laps = laps['Driver'].value_counts().reset_index()
total_laps.columns = ['Driver', 'TotalLaps']

standings = results[['Abbreviation', 'Position']].rename(columns={'Abbreviation': 'Driver'})

final_df = pd.merge(standings, total_laps, on='Driver', how='left')
final_df = final_df[['Driver', 'TotalLaps', 'Position']]

final_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Sort by Position
final_df_sorted = final_df.sort_values('Position')

# Step 2: Plot
plt.figure(figsize=(12, 6))
sns.barplot(
    data=final_df_sorted,
    x='Driver',
    y='TotalLaps',
    order=final_df_sorted['Driver'],  # Ensures order by position
    palette='viridis'
)
plt.title('Total Laps Completed by Each Driver (Ordered by Standings)')
plt.xlabel('Driver (Abbreviation)')
plt.ylabel('Total Laps')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
drivers = bahrain_session.drivers
print(drivers)
```

```python
drivers = [bahrain_session.get_driver(driver)["Abbreviation"] for driver in drivers]
print(drivers)
```

```python
stints = laps[["Driver", "Stint", "Compound", "LapNumber"]]
stints = stints.groupby(["Driver", "Stint", "Compound"])
stints = stints.count().reset_index()
```

```python
stints = stints.rename(columns={"LapNumber": "StintLength"})
print(stints)
```

```python
fig, ax = plt.subplots(figsize=(5, 10))

for driver in drivers:
    driver_stints = stints.loc[stints["Driver"] == driver]

    previous_stint_end = 0
    for idx, row in driver_stints.iterrows():
        # each row contains the compound name and stint length
        # we can use these information to draw horizontal bars
        compound_color = f1plt.get_compound_color(row["Compound"],
                                                            session=bahrain_session)
        plt.barh(
            y=driver,
            width=row["StintLength"],
            left=previous_stint_end,
            color=compound_color,
            edgecolor="black",
            fill=True
        )

        previous_stint_end += row["StintLength"]


plt.title("2022 Bahrain Grand Prix Strategies")
plt.xlabel("Lap Number")
plt.grid(False)
# invert the y-axis so drivers that finish higher are closer to the top
ax.invert_yaxis()

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.tight_layout()
plt.show()
```
