Based on the official records, here is the total number of races for each Formula 1 season from 2018 to 2024:

2018: 21 races

2019: 21 races

2020: 17 races (This season was shortened due to the COVID-19 pandemic)

2021: 22 races

2022: 22 races

2023: 23 races

2024: 24 races

So 24 possible tracks

```python
def get_race_data():
    race_data= []
    for i in range(24):
        session = f1.get_session(2024,i+1,'R')
        session.load(laps=True,weather=False,telemetry=False)

        race=session.event.Country
        total_laps=session.total_laps

        race_data.append({"Race" : race , "Total Laps" : total_laps })

    return pd.DataFrame(data=race_data,columns=['Race','Total Laps'])
```

```python
race_and_laps = get_race_data()
```
